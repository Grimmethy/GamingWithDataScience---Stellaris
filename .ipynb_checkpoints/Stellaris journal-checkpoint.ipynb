# Economy Calculator

```python
# Let's start with energy credit production.
# All production is in parts per month

# Storage:
# Base Capacity: 50,000
energy_credit_capacity = 50000

# Income:
# Base income: 50

# Current stations credit production: 29.7
# number of stations: 3
# production unmodified: 3 * 9
# machine intelligence modifier: +10%
station_credit_production_base = 27
station_credit_production_modifier = .10

station_credit_production = station_credit_production_base * (1 + station_credit_production_modifier)


# Current job credit production: 43.92
# Technician jobs: 6
# Job produces: 7.3
num_technicians = 6
technician_job_production_base = 7.30

job_credit_production = num_technicians * technician_job_production_base

# Arcane generator produces a base of 10 energy per month.
deposit_production = 10


energy_credit_revenue = 50 + station_credit_production + job_credit_production + deposit_production
## Base, Stations, Jobs, Deposits

# Expenses:

# Ships maintenance: 3.52
corvette_ship_upkeep = .3
# Components seem to follow a rule that the credit upkeep = 1% of the alloy cost to purchase.
dock_maintenance_bonus = .75
corvette_component_upkeep = .6
corvette_total_upkeep = corvette_ship_upkeep + corvette_component_upkeep
corvette_quantity = 0
corvette_quantity_docked = 3
corvette_docked_maintenance = corvette_total_upkeep * dock_maintenance_bonus

corvette_maintenance = corvette_quantity * corvette_total_upkeep + corvette_quantity_docked * corvette_docked_maintenance

science_ship_maintenance = .75
science_ship_quantity = 1
construction_ship_maintenance = .75
construction_ship_quantity = 1

utility_ship_maintenance = science_ship_maintenance * science_ship_quantity + construction_ship_maintenance * construction_ship_quantity

ship_energy_maintenance = corvette_maintenance + utility_ship_maintenance

# Station Maintenance: 3
station_quantity = 3
station_energy_maintenance_base = 1
station_energy_maintenance = station_quantity * station_energy_maintenance_base

# Starbase Maintenance: 4
starbase_energy_cost_base = 2
shipyard_energy_cost = 1
service_umbilicals_cost = 1
starbase_total_energy_cost = starbase_energy_cost_base + shipyard_energy_cost + service_umbilicals_cost

# Buildings maintenance cost: 15
admin_array = 5
mach_assembly = 2
uplink_node = 2
research_lab = 2
alloy_foundry = 2
hydro_farm = 2

total_building_maintenance = admin_array + mach_assembly + uplink_node + research_lab + alloy_foundry + hydro_farm

# Districts maintenance cost: 10
nexus_segment_energy_cost = 10
nexus_segment_quantity = 1

district_energy_maintenance = nexus_segment_quantity * nexus_segment_energy_cost

# Jobs maintenance cost: 16

coordinator_energy_maintenance = 4
calculator_energy_maintenance = 4

coordinator_quantity = 2
calculator_quantity = 2

total_job_energy_maintenance = coordinator_quantity * coordinator_energy_maintenance + calculator_quantity * calculator_energy_maintenance


# Pops energy maintenance: 19

machine_pop_energy_maintenance = 1
machine_pop_quantity = 19

pop_energy_maintenance = machine_pop_quantity * machine_pop_energy_maintenance

# Leaders energy maintenance: 10
leader_energy_maintenance = 2
leader_quantity = 5

total_leader_energy_maintenance = leader_quantity * leader_energy_maintenance

energy_credit_expense = ship_energy_maintenance + station_energy_maintenance + starbase_total_energy_cost + total_building_maintenance + district_energy_maintenance + total_job_energy_maintenance + pop_energy_maintenance + total_leader_energy_maintenance

## Ships, Stations, Starbases, Buildings, Districts, Jobs, Pops, Leaders

print(energy_credit_revenue)
print(energy_credit_expense)
```
